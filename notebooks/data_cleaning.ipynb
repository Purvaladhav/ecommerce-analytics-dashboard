```python
import pandas as pd

df = pd.read_csv('../data/processed/cleaned_data.csv')

# Basic overview
df.head()
df.info()

# Total revenue
df['revenue'].sum()

# Monthly revenue
df.groupby('month')['revenue'].sum()

# Top product categories
df.groupby('product_category_name')['revenue'].sum().sort_values(ascending=False).head(10)

# Top cities
df.groupby('customer_city')['revenue'].sum().sort_values(ascending=False).head(10)

# Order status
df['order_status'].value_counts()
```
